# Loading the Libraries and Dataset

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings("ignore")

df = pd.read_csv( 'IPL.csv')

df.head()
```

**Now, Here comes some Basic Questions**

**1.Which team won the most matches?**

```python
# Match wins
match_wins = df['match_winner'].value_counts()
match_wins
```

```python
#barplot of seaborn library
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
plt.title("Most match win by team ")
```

**2.Toss decision trend**

```python
sns.countplot(x = df['toss_decision'],palette='rainbow')
plt.title("Toss Decision Trends ")
```

**3.Toss Winner vs Match Winner**

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count *100)/df.shape[0]
percentage.round(2)
```

**4.How do teams win? (Runs vs Wickets)**

```python
sns.countplot(x = df['won_by'])
plt.title("Won by")
```

# Key Player Performances

**5. Most "Player of the Match" Awards**

```python
count = df['player_of_the_match'].value_counts().head(10)
```

```python
count
```

```python

sns.barplot(x = count.values , y = count.index,palette='mako')
plt.title("Top 10 plauers with man of the match")
```

**6.Top Scorers**

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

**7.Best bowling figures**

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```

**8.Largest victory Margins**

```python
# Largest victory margins
largest_margins = df.sort_values('margin', ascending=False).head(5)
print(largest_margins[['match_winner','margin','won_by']])
```

**9.Correlation between first and second innings scores**

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='first_ings_score', y='second_ings_score', data=df, color='teal')
plt.title('1st vs 2nd Innings Scores')
plt.xlabel('First Innings Score')
plt.ylabel('Second Innings Score')
plt.show()

corr = df[['first_ings_score','second_ings_score']].corr()
print(corr)
```

**10.Which team scored the highest single first innings total, and what was that score?**

```python
# Find the maximum first innings score
max_score = df['first_ings_score'].max()

# Find rows where that score occurred
highest_scores = df[df['first_ings_score'] == max_score]

# Display result
print("Highest first innings score:", max_score)
print(highest_scores[['team1', 'team2', 'venue', 'first_ings_score']])
```
